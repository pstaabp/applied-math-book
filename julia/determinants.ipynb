```julia
while true
A = rand(-4:4,3,3)
abs(det(A)) < 6 && break
end
A
```

```julia
det(A)
```

```julia
minor(A, i, j) = A[setdiff(1:size(A,1), i),setdiff(1:size(A,2),j)]
cofactor(A, i, j) = (-1)^(i+j) * det(minor(A,i,j))
```

```julia
minor(A,1,2)
```

```julia
minor(A,3,3)
```

```julia
cofactor(A,1,2),cofactor(A,3,3)
```

```julia
C = zeros(Int, 3,3)
for i=1:3
  for j=1:3
    C[i,j] = round(Int, cofactor(A,i,j))
  end
end
C
```

```julia
C'*A
```
